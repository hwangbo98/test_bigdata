```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('archive/basic2.csv', parse_dates=['Date'])
df.head()
```

```python
df.info()
```

```python
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month
df['day'] = df['Date'].dt.day
df['dayofweek'] = df['Date'].dt.dayofweek

df.info()
```

```python
df.head()
```

```python
df['weekend'] = df['dayofweek'].apply(lambda x: x>=5)

df.head()
```

```python
weekend_cond = (df['year']==2022) & (df['month']==5) & (df['weekend']==True)
weekday_cond = (df['year']==2022) & (df['month']==5) & (df['weekend']==False)

```

```python
weekend_sales = df[weekend_cond]['Sales'].mean()
weekday_sales = df[weekday_cond]['Sales'].mean()

print(round(abs(weekend_sales - weekday_sales),1))
```
